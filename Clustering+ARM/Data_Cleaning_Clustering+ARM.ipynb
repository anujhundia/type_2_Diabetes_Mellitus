```python
# load new_file.csv using pandas

import pandas as pd
df = pd.read_csv('/content/new_file.csv')
```

```python
medicine=df[['digestive_drugs','antiobesity','other_digestive','hematologic_drugs','cardiovascular_drugs','lipid_modifying','dermatologic_drugs',
             'gynecologic_drugs','sex_hormones','systemic_hormones','glucagon','calcium_homeostasis_drugs','antiinfectives','vaccines','antineoplastics',
             'endocrin_drugs','immunostimulants','immunosupresants','musculosceletal_drugs','anesthetics','analgesics','antiepileptics','antiparkinson',
             'antipsychotics','anxiolytics','pshycoanaleptics','other_nervous_drugs','antiparasitic','respiratory_sys_drugs','eye_ear_drugs','various_drugs']]
```

```python
diseases=df[['infectious_diseases','Malign_neoplasms','Obesity','Thyroid_dis','neoplasms_unknown','anemia','vitamin_deficiency','lipoprotein_met_dis','hematologic_dis',
       'endocrine_other','bipolar_affective_dis','depression','anxiety_dis','Other_mental_dis','neuropathies','diabetic_nueropathy','nervous_sys_dis',
       'cataract','retinopathy','refraction_dis','impacted_cerumen','tinnitus_h93_1','eye_other','otitis_externa_h60','mastoid_h60_h95','hypertension_i10',
       'ischemic_heart_dis','cardiomyopathies','cerebrovascular','other_circulatory','respiratory_sys','oral_dis','gastro_oes_reflux','dyspepsia','digestive_sys_dis',
       'skin_dis','musculoskeletal_dis','nephropaties','kidney_failure','other_genitourinary','pregnancy','birth','ceserian_multiple','other_pregnancy']]
```

```python
for column in medicine.columns:
    # Replace the value 1 in each column with a custom string
    medicine[column] = medicine[column].replace(1, column)

medicine.head(5)
```

```python
for column in diseases.columns:
    # Replace the value 1 in each column with a custom string
    diseases[column] = diseases[column].replace(1, column)

diseases.head(5)
```

```python
# Convert the DataFrame to transaction data
transactions_medicine = []
for _, row in medicine.iterrows():
    transaction = [col for col in row if col != 0]
    transactions_medicine.append(transaction)
```

```python
# Convert the DataFrame to transaction data
transactions_diseases = []
for _, row in diseases.iterrows():
    transaction = [col for col in row if col != 0]
    transactions_diseases.append(transaction)
```

```python
# Convert the list of transactions to a DataFrame
df_medicine = pd.DataFrame(transactions_medicine)

# Save the DataFrame to a CSV file
df_medicine.to_csv('medicine_arm.csv', header=False)
```

```python
# Convert the list of transactions to a DataFrame
df_diseases= pd.DataFrame(transactions_diseases)

# Save the DataFrame to a CSV file
df_diseases.to_csv('diseases_arm.csv', header=False)
```
